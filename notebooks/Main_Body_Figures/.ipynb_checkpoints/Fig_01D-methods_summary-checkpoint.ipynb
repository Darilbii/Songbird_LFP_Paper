```python
def event_shape_correction_visualizing(chunk_events):
    """ Reshape the output of get_event_related_nd_chunk to be shape of [Instances]->(Channels, Samples)"""
    corrected = []
    
    check = min([len(np.shape(i)) for i in chunk_events])
    
    
    if check == 2:
        for chunk in chunk_events:
            if len(chunk.shape) == 3:
                for instances in chunk:
                    corrected.append(instances)
            else:
                corrected.append(chunk)
    if check == 3:
        for chunk in chunk_events:
            if len(chunk.shape) == 4:
                for instances in chunk:
                    corrected.append(instances)
            else:
                corrected.append(chunk)
    return corrected
```
